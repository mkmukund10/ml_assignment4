```python
import mdptoolbox, mdptoolbox.example
import numpy as np
import gym
from gym import wrappers
import time
import sys
import matplotlib.pyplot as plt
import hiive
from hiive.mdptoolbox import mdp
from itertools import chain
from mdp import QLearning
```

```python
print('Q LEARNING WITH FOREST MANAGEMENT')
np.random.seed(0)
P, R = mdptoolbox.example.forest(S=2000,p=0.01)
value_f = []
policy = []
iters = []
time_array = []
Q_table = []
rew_array = []
for epsilon in [0.05,0.15,0.25,0.5,0.75,0.95]:
    st = time.time()
    pi = QLearning(P,R,0.95, n_iter=100000) #, epsilon=epsilon, gamma=0.95)
    pi.run(epsilon)
    end = time.time()
    rew_array.append(pi.reward_array)
    value_f.append(np.mean(pi.V))
    #print ((pi.reward_array))
    policy.append(pi.policy)
    time_array.append(end-st)
    Q_table.append(pi.Q)
```

```python
plt.plot(range(0,100000), rew_array[0],label='epsilon=0.05')
plt.plot(range(0,100000), rew_array[1],label='epsilon=0.15')
plt.plot(range(0,100000), rew_array[2],label='epsilon=0.25')
plt.plot(range(0,100000), rew_array[3],label='epsilon=0.50')
plt.plot(range(0,100000), rew_array[4],label='epsilon=0.75')
plt.plot(range(0,100000), rew_array[5],label='epsilon=0.95')
plt.legend()
plt.xlabel('Iterations')
plt.grid()
plt.title('Forest Management-QLearning-Decaying Epsilon @ Gamma=0.95')
plt.ylabel('Average Reward')
plt.show()

# plt.subplot(1,6,1)
# plt.imshow(Q_table[0][:20,:])
# plt.title('E=0.05')

# plt.subplot(1,6,2)
# plt.title('E=0.15')
# plt.imshow(Q_table[1][:20,:])

# plt.subplot(1,6,3)
# plt.title('E=0.25')
# plt.imshow(Q_table[2][:20,:])

# plt.subplot(1,6,4)
# plt.title('E=0.50')
# plt.imshow(Q_table[3][:20,:])

# plt.subplot(1,6,5)
# plt.title('E=0.75')
# plt.imshow(Q_table[4][:20,:])

# plt.subplot(1,6,6)
# plt.title('E=0.95')
# plt.imshow(Q_table[5][:20,:])
# plt.colorbar()
# plt.show()
```

```python
print('Q LEARNING WITH FOREST MANAGEMENT')
np.random.seed(0)
P, R = mdptoolbox.example.forest(S=2000,p=0.01)
value_f = []
policy = []
iters = []
time_array = []
Q_table = []
rew_array = []
for epsilon in [0.05,0.15,0.25,0.5,0.75,0.95]:
    st = time.time()
    pi = hiive.mdptoolbox.mdp.QLearning(P,R,gamma=0.9,epsilon=epsilon, alpha=0.9)
    pi.run()
    end = time.time()
    rew_array.append(list(chain.from_iterable(pi.v_mean)))
    value_f.append(np.mean(pi.V))
    #print (len(pi.v_mean))
    #print (pi.v_mean)
    policy.append(pi.policy)
    time_array.append(end-st)
    Q_table.append(pi.Q)
#print((rew_array[0]))
```

```python
plt.plot([0.05,0.15,0.25,0.5,0.75,0.95],time_array)
plt.xlabel('Epsilon Values')
plt.grid()
plt.title('Forest Management-QLearning-Decaying Epsilon @ Gamma=0.9')
plt.ylabel('Execution Time (s)')
plt.show()
```
