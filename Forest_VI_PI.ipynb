```python
import gym
import time
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from gym.envs.toy_text.frozen_lake import generate_random_map
from hiive.mdptoolbox.mdp import PolicyIteration, ValueIteration, QLearning, PolicyIterationModified
from hiive.mdptoolbox.example import forest
from matplotlib import pyplot as ply
```

```python
print('POLICY ITERATION WITH FOREST MANAGEMENT')
np.random.seed(0)
P, R = forest(S=2000,p=0.01)
pi_value_f = [0]*10
pi_policy = [0]*10
pi_iters = [0]*10
pi_time_array = [0]*10
pi_gamma_arr = [0] * 10

vi_value_f = [0]*10
vi_policy = [0]*10
vi_iters = [0]*10
vi_time_array = [0]*10
vi_gamma_arr = [0] * 10

for i in range(0,10):
    pi = PolicyIterationModified(P, R, (i+0.5)/10)
    pi.run()
    pi_gamma_arr[i]=(i+0.5)/10
    pi_value_f[i] = np.amax(pi.V)
    pi_policy[i] = pi.policy
    pi_iters[i] = pi.iter
    pi_time_array[i] = pi.time
    
    vi = ValueIteration(P, R, (i+0.5)/10)
    vi.run()
    vi_gamma_arr[i]=(i+0.5)/10
    #print (len(pi.V))
    vi_value_f[i] = np.amax(vi.V)
    #print(pi.V)
    vi_policy[i] = vi.policy
    vi_iters[i] = vi.iter
    #print (pi.iter)
    vi_time_array[i] = vi.time
```

```python
ply.plot(pi_gamma_arr,pi_iters, label='PI')
ply.plot(vi_gamma_arr,vi_iters, label='VI')
ply.xlabel('Gammas')
ply.ylabel('Iterations to Converge')
ply.title('Forest Management - PI/VI - Convergence Analysis')
ply.legend(loc = 'upper left')
ply.grid()
ply.show()
```

```python
plt.plot(pi_gamma_arr,pi_time_array, label = 'PI')
plt.plot(vi_gamma_arr,vi_time_array, label = 'VI')
plt.xlabel('Gammas')
plt.ylabel('Time')
plt.title('Forest Management - PI/VI - Convergence Analysis')
plt.grid()
ply.legend(loc = 'upper left')
plt.show()
```

```python
plt.plot(pi_gamma_arr,pi_value_f, label='PI')
plt.plot(vi_gamma_arr,vi_value_f, label='VI')
plt.xlabel('Gammas')
plt.ylabel('MaxV')
plt.title('Forest Management - PI/VI - Convergence Analysis')
ply.legend(loc = 'upper left')
plt.grid()
plt.show()
```

```python
np.random.seed(0)
sizes = np.arange(10, 2011, 200)
pi_value_f = []
pi_policy = []
pi_iters = []
pi_time_array = []
pi_gamma_arr = []

vi_value_f = []
vi_policy = []
vi_iters = []
vi_time_array = []
vi_gamma_arr = []

for i in sizes:
    P, R = forest(S=i,p=0.01)
    pi = PolicyIterationModified(P, R, 0.99)
    pi.run()
    pi_gamma_arr.append(i)
    pi_value_f.append(np.amax(pi.V))
    pi_policy.append(pi.policy)
    pi_iters.append(pi.iter)
    pi_time_array.append(pi.time)
    
    vi = ValueIteration(P, R, 0.99)
    vi.run()
    vi_gamma_arr.append(i)
    #print (len(pi.V))
    vi_value_f.append(np.amax(vi.V))
    #print(pi.V)
    vi_policy.append(vi.policy)
    vi_iters.append(vi.iter)
    #print (pi.iter)
    vi_time_array.append(vi.time)
```

```python
plt.plot(pi_gamma_arr,pi_time_array, label = 'PI')
plt.plot(vi_gamma_arr,vi_time_array, label = 'VI')
plt.xlabel('Size')
plt.ylabel('Time')
plt.title('Forest Management - PI/VI Time Vs Size - Convergence Analysis')
plt.grid()
ply.legend(loc = 'upper left')
plt.show()
```
